# Line Chart

```python
# Line Chart
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6,7,8]
y = [10,3,5,6,7,3,2,8]
plt.plot(x,y)
plt.title("Line Chart")
plt.xlabel("day #")
plt.ylabel("number of students")
```

# Bar Chart

```python
import matplotlib.pyplot as plt
catergories = ["A", "B" , "C", "D", "E"]
values = [25, 45, 56, 46, 78]
plt.bar(catergories, values)
plt.title("Bar Chart")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

# Pie Chart

```python
import matplotlib.pyplot as plt
labels = ["Category A", "Category B", "Category C"]
sizes = [35, 40, 20]
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title("Pie Chart")
plt.show()
```

# Histogram

```python
import matplotlib.pyplot as plt
import numpy as np

data = np.random.randn(1000)
plt.hist(data, bins=30, edgecolor='black')
plt.title("Histogram")
plt.xlabel("values")
plt.ylabel("Frequency")
plt.show()
```

# Scatter Plot

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.rand(50)
y = 2 * x +1 + 0.1 * np.random.rand(50)
plt.scatter(x, y)
```

# Heatmap

```python
import seaborn as sns
import numpy as np 

data = np.random.rand(10, 10)

sns.heatmap(data, annot=True)
plt.title("Heatmap")
plt.show()
```

# Area Chart


```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y1 = [10,14, 25, 30, 35]
y2 = [5,10,20,25,30]

plt.fill_between(x, y1, y2, color='skyblue', alpha=0.4)
plt.show()
```

# Activation function  for neural network

```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(arr, scale=1):
    arr = np.asarray(arr)
    result = 1/(1 + np.exp(-arr*scale))
    return result

x = np.linspace(-5, 5)
y = sigmoid(x)

print(y)

fig, ax = plt.subplots()
ax.plot(x, y)
```

```python
import numpy as np
import seaborn as sns

# Define the sigmoid function
def sigmoid(arr, scale=1):
    return 1 / (1 + np.exp(-arr*scale))

# Create an array of x values
x = np.linspace(-5, 5, 100)

# Calculate the corresponding y values
y = sigmoid(x)

# Create the plot
sns.lineplot(x=x, y=y)
```
